RECURSIVE FEATURE ELIMINATION

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#no of features
nof_list = np.arange(1,30)            
high_score = 0

#Variable to store the optimum features
nof = 0    
      
score_list =[]
for n in range(len(nof_list)):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 3534530)

    model = LinearRegression()
    rfe = RFE(model, step=nof_list[n])

    X_train_rfe = rfe.fit_transform(X_train,y_train)
    X_test_rfe = rfe.transform(X_test)

    model.fit(X_train_rfe,y_train)
    
    score = model.score(X_test_rfe,y_test)
    score_list.append(score)

    print("Obtained Score: ", score, "  ||  High Score: ", high_score )
    if(score >= high_score):
        high_score = score
        nof = nof_list[n]

print("Optimum number of features: %d" %nof)
print("Score with %d features: %f" % (nof, high_score))
```

WEIGHTED K-MEANS

```python
import math
```

```python
class wKmeans:
    
    def __init__(self, k, X_train, beta = 2, max_iter=10000, weight_init='uniform'):
        self.k = k
        self.beta = beta
        self.MAX_ITER = max_iter
        self.X_train = X_train
        
        self.u = [[0]*k]*self.X_train.shape[0]
        self.u = np.array(self.u, dtype='float64')
        
        self.z = []
        for i in range(k):
            self.z.append(self.X_train.iloc[[i]].to_numpy())
        self.z = np.array(self.z, dtype='float64')

        self.w = [1.0/self.X_train.shape[1]]*self.X_train.shape[1]
        self.w = np.array(self.w, dtype='float64')
        
        self.weight_init = weight_init
        
        self.prev_u = np.array([])
        self.prev_z = np.array([])
        self.prev_w = np.array([])
        self.categorical = np.array([])
    
    def linear_dist(self, p1, p2):
        return p1-p2
    
    def distance_function(self, p1, p2):
        p1 = p1.to_numpy()[0]
        p2 = p2.to_numpy()[0]
        sum = 0.0
        i = 0
        for coord in zip(p1, p2):
            # If its a categorical feature
            if i in self.categorical:
                if coord[0] != coord[1]:
                    sum += ((self.w[i]**self.beta))**2
            # If its a numerical feature
            else:
                sum += ((self.w[i]**self.beta)*self.linear_dist(coord[0],coord[1]))**2
            i += 1
        return math.sqrt(sum)
    
    def getDj(self, j):
        Dj = 0
        for m in range(self.k):
            for n in range(self.X_train.shape[0]):
                # If its a categorical feature
                if j in self.categorical:
                    if (self.X_train.iloc[[n]].to_numpy()[0])[j] != self.z[m][0][j]:
                        Dj += 1
                # If its a numerical feature
                else:
                    Dj += self.u[n][m]*((self.X_train.iloc[[n]].to_numpy()[0])[j] - self.z[m][0][j])**2
        return Dj
    
    def fit(self, categorical):
        self.categorical = categorical
        a = 0
        while True:
            a = a + 1
            for i in range(self.X_train.shape[0]):
                distances = []
                for j in range(self.k):
                    distances.append(self.distance_function(self.X_train.iloc[[i]], pd.DataFrame(self.z[j])))
                min_cluster = np.argmin(distances)
                for j in range(self.k):
                    if(j == min_cluster):
                        self.u[i][j] = 1
                    else:
                        self.u[i][j] = 0

            # Convergence
            if np.array_equal(self.u, self.prev_u):
                print("Convergence by datapoints")
                break
            else:
                self.prev_u = self.u

            # Update the centroids
            for i in range(self.k):
                self.z[i] = np.array([0]*self.X_train.shape[1])
                n = 0
                for j in range(self.X_train.shape[0]):
                    # If the current row belongs to this cluster
                    if(self.u[j][i] == 1):
                        # If its a categorical feature
                        if j in self.categorical:
                            count = dict()
                            for a in range(self.X_train.shape[0]):
                                if(self.u[a][i] == 1):
                                    count[self.X_train.iloc[[a]].to_numpy()[0][j]] = count.get(self.X_train.iloc[[a]].to_numpy()[0][j], 0) + 1
                            mode = max(count, key=count.get)
                            self.z[i] = mode
                        # If its a numerical feature
                        else:
                            self.z[i] += self.X_train.iloc[[j]].to_numpy()[0]
                        n += 1
                            
                # Avoid dividing by 0 
                if n == 0:
                    n = 1
                self.z[i] = np.divide(self.z[i], n)

            # Convergence
            if np.array_equal(self.z, self.prev_z):
                print("Convergence by centroids")
                break
            else:
                self.prev_z = self.z


            # Update the weights
            if(self.beta == 1):
                j_min = float('inf')
                j_min_index = 0
                for j in range(len(self.w)):
                    Dj = getDj(j)
                    if(Dj < j_min):
                        j_min = Dj
                        j_min_index = j
                for j in range(len(self.w)):
                    self.w[j] = 0
                self.w[j_min_index] = 1        
            else:
                for j in range(len(self.w)):
                    Dj = self.getDj(j)
                    if Dj == 0:
                        self.w[j] = 0
                    else:
                        denom = 0
                        for j1 in range(self.X_train.shape[1]):
                            Dt = self.getDj(j1)
                            if(Dt != 0):
                                denom += (Dj/Dt)**(1/(self.beta - 1))
                        self.w[j] = 1.0/denom

            # Convergence
            if np.array_equal(self.w, self.prev_w):
                print("Convergence by weights")
                break
            else:
                self.prev_w = self.w                    

            # Max iterations reached?
            if a >= self.MAX_ITER:
                print("MAX_ITER reached")
                break
        
        print("Took {} iterations to converge".format(a))
        
    def getWeights(self):
        return self.w
```
